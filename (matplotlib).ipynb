```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y1=[50,51,52,48,47,49,46]
y2=[43,42,40,44,33,35,37]
y3=[45,48,48,46,40,42,41]
```

```python
plt.figure(figsize=(4,4))

plt.xlabel("title1")
plt.ylabel('title2')
plt.title('title3')

plt.scatter(x,y1, label="y1")
plt.scatter(x,y2, label="y2")
plt.scatter(x,y3, label="y3")

plt.legend()

plt.xticks(rotation=90)

plt.grid()
```
