```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data)
df['target'] = digits.target
df
```

```python
x=df.drop('target',axis=1)
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split  as aa
mm=aa(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier as ss
model=ss().fit(mm[0],mm[2])
```

```python
model.score(mm[1],mm[3])
```

```python
mm[3]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(model.predict(mm[1]),mm[3])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(15,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
